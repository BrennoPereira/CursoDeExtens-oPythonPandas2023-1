```python
#importando biblioteca pandas no Python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Usar print para ver o datafarme ! (Mostra tudo)

print(combustiveis_df)
```

```python
# Mostrar as primeiras 5 linhas 
# Display e print quase mesma coisa, porém o display mostra mais organizado.
# função .head filtra o numeros de linhas

display(combustiveis_df.head())
```

```python
# Exibir apenas as 15 primeiras linhas.

display(combustiveis_df.head(15))
```

```python
# Comandos Dataframe.shape linhas e colunas e 
# Dataframe.describe() mostra estatisticas "mais básicas" que a gente precisa
display(combustiveis_df.shape[0])
print(combustiveis_df.describe())
```

```python
# Info: Quais são colunas e que tipo de dados cada um tem... *importante*
display(combustiveis_df.info())
```

```python
# Filtrar apenas uma coluna e as 250 primeiras linhas
print(combustiveis_df.head(250)["Data da Coleta"])
```

```python
# Filtrar apenas uma coluna e as 50 ultimas linhas, Usando o .tail 
print(combustiveis_df.tail(50)["Revenda"])
```

```python
# Criando novo dataframe apenas com algumas colunas que eu quero;
simples_df = combustiveis_df[['Revenda','Municipio','Valor de Venda','Produto']]
display(simples_df.tail())
```

```python
# Exibir a 4a linha  ---> só as imformações que quero.
display(simples_df.loc[4][["Revenda",'Valor de Venda']])
```

```python
# Exibir da 10a linha até a 20a linha. 
display(simples_df.loc[10:20])
#display(simples_df.loc[10:20][['Revenda','Valor de Venda','Produto']])
```

```python
# Criar um dataframe gas_df contendo 
# apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
# somente com combustível sendo GASOLINA e exibir na tela

gas_df = simples_df.loc[simples_df['Produto'] == 'GASOLINA']
display(gas_df)
print(gas_df['Valor de Venda'].max())
```

```python
# Exibir qual é o posto e munícipio da gasolina mais cara do brasil e outro produto.
display(combustiveis_df[['Valor de Venda','Revenda','Municipio','Produto']].max() )
display(gas_df[['Revenda','Municipio','Valor de Venda']].max())
```

```python
# Dataframe.loc[] com múltiplas condições para filtragem. 
# Quais são os preços, postos que vendem ETANOL na minha cidade (SÃO PAULO)
etanol_saopaulo_df = simples_df.loc[(simples_df['Produto'] == 'ETANOL')  & (simples_df['Municipio'] == 'SAO PAULO')]
#etanol_indaiatuba_df = simples_df.loc[(simples_df['Produto'] == 'ETANOL') & (simples_df['Municipio'] == 'INDAIATUBA')] 
display(etanol_saopaulo_df)
etanol_saopaulo_df.sort_values(by='Valor de Venda', inplace=True)
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
#Passo anterior usando .isin
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL Brasil todo?
# Round para menos casa decimal. 
media_por_combustivel_df = simples_df[['Produto', 'Valor de Venda',]].groupby(by='Produto').sum().round(2)
display(media_por_combustivel_df)
```

```python
#Quero adicionar uma coluna de valor booleano no combustiveis_df
# chamada "Ativo" que, por padrão, vai ser True para 
# TODAS as linhas
combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())
```
